**Problem Statement:**

Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

**Data Set used :-** #https://www.kaggle.com/datasets/thedevastator/analyzing-credit-card-spending-habits-in-india/data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data from CSV file
df = pd.read_csv('Credit card transactions - India - Simple.csv')
df
```

```python
df.head(4000)
```

```python
# Bar chart for the distribution of a categorical variable (gender)
plt.figure(figsize=(6, 5))
sns.set(style="whitegrid")
sns.countplot(x='Gender', data=df.count, palette='dark')
plt.title('Distribution of Gender in the Top 4000 Rows')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

From the dataset we can say that Female has more Credit card transactions compare to male
